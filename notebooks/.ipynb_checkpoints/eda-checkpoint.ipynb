## Exploratory Data Analysis

The goal of this notebook is to perform some initial exploratory data analysis to gather a better understanding of the dataset. In order to achieve a better understanding univariate, bivariate and multivariate analysis is to be performed. Since there is no target variable in our dataset we would need to explore the interactions between our products. The end goal is to create a unsupervised recommendation algorithm so we must perform the necessary analysis to get us closer to our end goal. 

### Imports & Data

```python
#imports needed for this project
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#data reading
df = pd.read_csv("../data/train.csv")
df.head()
```

### Understanding Our Distributions

Univariate analysis to get a better understanding of how our data is distributed. 

```python
fig, ax = plt.subplots(figsize=(8, 6))
#count plot of the genre column
plt.title('Count plot of the genre column')
sns.countplot(x='genre', data=df, ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
#count plot of the topic column
plt.title('Count plot of the topic column')
sns.countplot(x='topic', data=df, ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(12, 10))
#countplot of the release date column
plt.title('Count plot of the release_date column')
sns.countplot(x='release_date', data=df, ax=ax)
plt.xticks(rotation=70)
plt.tight_layout()
```

```python
#distribution of our len column
sns.displot(df, x='len', height=10).set(title='Distribution of our len column')
```

```python
#distribution of our age column
sns.displot(df, x='age', height=8).set(title='Distribution of our age column')
```

Key takeaways from our univariate distributions
- The **pop** genre is the most common genre while the **hip hop** genre is the least common
- The **sadness** topic is the most common while **feelings** is the least common topic
- We can also see that most of the songs in our dataset are newer while songs from 1950-1956 are less common in our dataset
- The distribution of the **lengths** of our songs seems right skewed with our median around 50
- The distribution of the **age** of our songs is nearly uniform except there is a larger amount of songs that are between 0.0 - 0.1 of age, this shows us that most of the songs are relatively new

Now we can get a better look at how our columns relate to each other using bivariate analysis

```python
#box plots showing distributions of genre and topic columns by our len column
fig, ax = plt.subplots(1, 2, figsize=(15, 10))
sns.boxplot(data=df, x='len', y='genre', ax=ax[0], showfliers=False)
sns.boxplot(data=df, x='len', y='topic', ax=ax[1], showfliers=False)
```

```python
#box plots showing distributions of genre and topic columns by our age column
fig, ax = plt.subplots(1, 2, figsize=(15, 10))
sns.boxplot(data=df, x='age', y='genre', ax=ax[0], showfliers=False)
sns.boxplot(data=df, x='age', y='topic', ax=ax[1], showfliers=False)
```

```python
#add a decades column to our dataframe
decades = ['50s', '60s', '70s', '80s', '90s', '00s', '10s']
df['decade'] = pd.cut(df['release_date'], 7, labels=decades)
```

```python
#here we create a pivot table counting genre by decade
genre_pivot = df.pivot_table(index=['decade'], columns='genre', aggfunc='size')
genre_pivot
```

```python
#now that we have our pivot table it would be nice to visualize it
ax = genre_pivot.plot(kind='bar', width=0.65)
#change the parameters of our figure for better visualization
fig = ax.get_figure()
fig.set_size_inches(12, 10)
#set the title of our figure
ax.set_title('Bar plot displaying genre counts by decade')
#show our figure
plt.show()
```

```python
#now we can do the same thing but instead we can visualize topic by decade
topic_pivot = df.pivot_table(index=['decade'], columns='topic', aggfunc='size')
topic_pivot
```

```python
#now that we have our pivot table it would be nice to visualize it
ax = topic_pivot.plot(kind='bar', width=0.7)
#change the parameters of our figure for better visualization
fig = ax.get_figure()
fig.set_size_inches(12, 10)
#set the title of our figure
ax.set_title('Bar plot displaying topic counts by decade')
#show our figure
plt.show()
```

Key takeaways from our bivariate distributions
- It seems that **rock, country and blues** are the genres with least length while **obscene** sticks out as a topic that has the longest length
- We can also see that **hip hop** is the genre with the smallest distribution of age and **obscene** is the topic that has the smallest distribution of age, this could be useful to see how hip hop and obscene are related to each other
- We can see that the **pop** genre is the most prominent genre in every decade except the 1950s
- Another interesting observation is the fact that the **obscene** topic skyrockets from the 2000s decade to the 2010s decade

```python
#create a correlation heatmap of our numerical columns
plt.figure(figsize=(16, 8))
sns.heatmap(
    df.select_dtypes(include=np.number).iloc[:, 1:].corr(),
    vmin=-1,
    vmax=1,
    annot=True,
    cmap='BrBG'
)
```

Key takeaways from our multivariate distribution
- We can see that **age** and **release_date** are negatively correlated, which makes sense because as our songs release date approaches 1950 the same songs age will approach 1
    - This tells us that we should get rid of one of this columns in our model since they are directly correlated
- We can also see that **len** and **onbscene** have a correlation of 0.44 which tells us that the more obscene our song is it tends to be longer in length
